```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
PATH = "../supervised/Agriculture-Vision-2021/train/images/rgb"
```

```python
imgs = os.listdir(PATH) # images path
```

```python
ids = set([x.split("_")[0] for x in imgs])
```

```python
fields = dict()
```

```python
for i in ids:
    fields[i] = []
    for img in imgs:
        if i == img.split("_")[0]:
            fields[i].append(img)
```

```python
def show_img(filename):
    plt.imshow(cv2.imread(PATH + filename, cv2.IMREAD_COLOR))
```
